```python
import pandas as pd

# Load the two CSV files
file_a = r"C:\Users\990887\Downloads\cases_export_2025-05-07.csv"  # Your file path
file_b = r"C:\Users\990887\Downloads\cases_export_2025-05-06.csv"  # Your file path

df_a = pd.read_csv(file_a)
df_b = pd.read_csv(file_b)

# Ensure both DataFrames have the same shape
if df_a.shape != df_b.shape:
    print("Error: The CSV files have different dimensions.")
    print(f"File A shape: {df_a.shape}, File B shape: {df_b.shape}")
    exit()

# Ensure both DataFrames have the same columns
if not df_a.columns.equals(df_b.columns):
    print("Error: The CSV files have different columns.")
    print(f"File A columns: {df_a.columns}")
    print(f"File B columns: {df_b.columns}")
    exit()

# Compare the two DataFrames
# Create a DataFrame to store differences
differences = df_a != df_b

# Initialize a list to store mismatch details
mismatch_details = []

# Iterate through each cell to find mismatches
for row in range(df_a.shape[0]):
    for col in df_a.columns:
        if differences.loc[row, col]:  # If the cell values differ
            # Handle NaN values (NaN != NaN in pandas, so we need special handling)
            val_a = df_a.loc[row, col]
            val_b = df_b.loc[row, col]
            if pd.isna(val_a) and pd.isna(val_b):
                continue  # Skip if both are NaN (not a meaningful difference)
            mismatch_details.append({
                'Row': row + 2,  # Adding 2 to account for header and 1-based indexing
                'Column': col,
                'Value in A': val_a,
                'Value in B': val_b
            })

# Convert mismatch details to a DataFrame for better visualization
mismatch_df = pd.DataFrame(mismatch_details)

# Output results
if mismatch_df.empty:
    print("No differences found between the two CSV files.")
else:
    print("Mismatches found:")
    print(mismatch_df)

# Optionally, save the differences to a new CSV or Excel file
mismatch_df.to_csv('differences_output.csv', index=False)
# If you prefer Excel output:
# mismatch_df.to_excel('differences_output.xlsx', index=False)
```
